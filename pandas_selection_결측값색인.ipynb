#3. Boolean Indexing: 조건을 활용한 색인
- Numpy의 Boolean 인덱싱과 같은 원리
- Boolean Index 로 받은 Index를 활용해서 True인 값만 색인해 낼 수 있음
* df[ ] 꺾쇠로 감싸주고, 그안에 Boolean Index를 넣어주면 됨

```python
# 검색조건

df['키'] > 180
```

```python
# df[검색조건]을 이런 형태로 사용 
df[df['키'] > 180]

#검색조건이 True가 되는 행만 출력됨
```

```python
df[df['혈액형']=='A']
```

위 방법이 매우 간편하지만, 모든 column을 출력해야한다는 한계 존재
###특정 column을 색인해 내고 싶은 경우는 어떻게?

####해결방법1. 맨 뒤에 출력할 column 붙이기

```python
df[df['키'] > 180 ]['이름']
```

```python
#'이름','키'를 2차원 리스트형태로 사용해야한다는 것 주의!!

df[df['키'] > 180 ][['이름', '키']]
```

```python
#키가 170보다 큰 사람의 혈액형만 컬럼으로 출력

df[df['키']>170]['혈액형']
```

####해결방법2. loc 활용

```python
df.loc[df['키'] > 180, '이름']
```

```python
df.loc[ df['키'] > 180, '이름': '성별']
```

```python
df.loc[ df['키'] > 180, ['이름', '키']]
```

```python
#강사님은 방법2 추천
```

#4. isin을 활용한 색인
- isin을 활용한 색인은 내가 조건을 걸고자 하는 값이 내가 정의한 list에 있을 때만 색인하려는 경우에 사용
- isin = 안에 내가 찾는 데이터가 있니?

```python
my_condition = ['플레디스', 'SM']
```

```python
# 검색조건 -> 소속사중에 플레디스 or SM이 있는지

df['소속사'].isin(my_condition)
```

```python
#내가 찾고자하는 소속사 아이돌의 행을 나타냄

df.loc[ df['소속사'].isin(my_condition) ]
```

```python
df.loc[ df['소속사'].isin(my_condition),'이름' ]
```

```python
df.loc[ df['소속사'].isin(my_condition), '소속사' ]
```

##5-2. 결측값 다루기

###Boolean indexing

```python
# 데이터프레임 전체에서 결측치가 있는지 확인할때 isna()

df.isna()

# 결측치이면 True, 아니면 False
```

isna( )를 활용해 전체적으로 어떤 column에 NaN값이 있는지 알 수 있고, 다음과 같은 방법으로 특정 column에서 빠진 값을 색출해 낼 수 있음

1. Boolean indexing으로 True가 return 되는 값이 NaN이라는 것을 알 수 있음

```python
# series에서 null값을 확인

df['그룹'].isnull()
```

2. NaN값만 색출해내기

```python
df['그룹'][df['그룹'].isnull()]
```

```python
df['그룹'][df['그룹'].isna()]
```

3. NaN이 아닌 값에 대하여 Boolean indexing

```python
df['그룹'].notnull()
```

4. NaN이 아닌 값만 색출해내기

```python
df['그룹'][df['그룹'].notnull()]
```

```python
# null이 아닌 데이터행만 골라내고, 그 중 '키'와 '혈액형' 컬럼만 출력

df.loc[ df['그룹'].notnull(), ['키', '혈액형'] ]
```
