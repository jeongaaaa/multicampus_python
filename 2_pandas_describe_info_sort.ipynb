# 5. 정렬하기

##7-1. index 기준으로 정렬

```python
#오름차순이 default

df.sort_index()
```

```python
#내림차순 정렬

df.sort_index(ascending=False)
```

##7-2. column 기준으로 정렬

```python
#오름차순 정렬

df.sort_values(by='키')

#Nan -> Not a Number 
#결측치가 있는 행의 경우 value기준으로 sorting하면 맨 끝으로 정렬됨
```

```python
#내림차순 정렬

df.sort_values(by='키', ascending=False)
```

```python
# +)함수 추가 옵션은 구글링해서 활용하기
```

###+) 복수정렬

```python
# 정렬 기준 1번 :'키', 키가 동일한 경우 정렬기준 2번

df.sort_values(by=['키', '브랜드평판지수'])

#오름차순
```

```python
df.sort_values(by=['키', '브랜드평판지수'], ascending=False)

#내림차순
```
